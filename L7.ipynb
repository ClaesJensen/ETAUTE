Opgave 1
Givet et stereosignal (altså to lydkanaler). Samplefrekvensen er 44.100 Hz.
a) Hvor stor datarate, målt i bits pr. sekund, har signalet, hvis hver sample er på 16 bit?
b) Og med 24 bit pr. sample?
Givet et lagermedie med en kapacitet på 650 MByte.
c) Hvor mange minutter er der plads til, hvis signalet er som i spørgsmål 1a?
Til sammenligning kigger vi nu på et signal i (fastnet-)telefonkvalitet, fs = 8 kHz, 8 bits pr. sample,
mono.
d) Hvor stor er dataraten af dette signal?
e) Mange stereolydfiler i MPEG-format er lagret med 128 kbit/s. Hvor stor en faktor er
datareduktionen i forhold til signalet i 1a?

```python
# Opgave 1

# Givet
fs = 44100                 # samplefrekvens i Hz
bits_per_sample_16 = 16    # bit pr sample
bits_per_sample_24 = 24
channels = 2               # stereo
storage_MB = 650           # lagerkapacitet i MB
fs_phone = 8000            # telefonkvalitet
bits_per_sample_phone = 8  # bit pr sample for telefon
channels_phone = 1         # mono
mpeg_rate_kbps = 128       # MPEG bitrate i kbit/s

# a) Datarate for 16 bit stereo
bitrate_16 = fs * bits_per_sample_16 * channels
print(f"a) Datarate (16 bit stereo): {bitrate_16:,} bit/s = {bitrate_16/1e6:.2f} Mbit/s")

# b) Datarate for 24 bit stereo
bitrate_24 = fs * bits_per_sample_24 * channels
print(f"b) Datarate (24 bit stereo): {bitrate_24:,} bit/s = {bitrate_24/1e6:.2f} Mbit/s")

# c) Hvor mange minutter på 650 MB (med 16 bit stereo)
storage_bytes = storage_MB * 1024**2           # konverter MB til bytes
storage_bits = storage_bytes * 8               # konverter bytes til bits
duration_seconds = storage_bits / bitrate_16   # sekunder lyd
duration_minutes = duration_seconds / 60
print(f"c) Varighed på 650 MB: {duration_minutes:.2f} minutter")

# d) Datarate for telefonkvalitet
bitrate_phone = fs_phone * bits_per_sample_phone * channels_phone
print(f"d) Datarate (telefonkvalitet): {bitrate_phone:,} bit/s = {bitrate_phone/1e3:.1f} kbit/s")

# e) Datareduktion ift. 16-bit stereo
reduction_factor = bitrate_16 / (mpeg_rate_kbps * 1000)
print(f"e) Datareduktion i forhold til 1a: ca. faktor {reduction_factor:.1f} gange")
```
